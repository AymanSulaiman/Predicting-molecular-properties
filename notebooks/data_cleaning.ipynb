```python
diapole_moments = pd.read_csv(os.path.join("..","data","dipole_moments.csv"))
magnetic_sheilding_tensors = pd.read_csv(os.path.join("..","data","magnetic_shielding_tensors.csv"))
mulliken_charges = pd.read_csv(os.path.join("..","data","mulliken_charges.csv"))
potential_energy = pd.read_csv(os.path.join("..","data","potential_energy.csv"))
scalar_coupling_contributions = pd.read_csv(os.path.join("..","data","scalar_coupling_contributions.csv"))
structures = pd.read_csv(os.path.join("..","data","structures.csv"))
```

```python
magnetic_sheilding_tensors_mulliken_charges_df = pd.merge(
    left=magnetic_sheilding_tensors,
    right=mulliken_charges,
    how="left",
    on=None,
    left_on=['molecule_name','atom_index'],
    right_on=['molecule_name','atom_index'],
    left_index=False,
    right_index=False,
    sort=True,
    suffixes=("_x", "_y"),
    copy=True,
    indicator=False,
    validate=None,
)
magnetic_sheilding_tensors_mulliken_charges_df.to_csv(os.path.join('..','data','joined_data','magnetic_sheilding_tensors_mulliken_charges_df.csv'), index=False)
```

```python
# magnetic_sheilding_tensors_mulliken_charges_df.head(2)
```

```python
diapole_moments_potential_energy_df = pd.merge(
    left=diapole_moments,
    right=potential_energy,
    how="inner",
    on=None,
    left_on='molecule_name',
    right_on='molecule_name',
    left_index=False,
    right_index=False,
    sort=True,
    suffixes=("_x", "_y"),
    copy=True,
    indicator=False,
    validate=None,
)
diapole_moments_potential_energy_df.to_csv(os.path.join('..','data','joined_data','diapole_moments_potential_energy_df.csv'), index=False)
```

```python
# structures
```

```python
magnetic_sheilding_tensors_mulliken_charges_structures_df = pd.merge(
    left=magnetic_sheilding_tensors_mulliken_charges_df,
    right=structures,
    how="left",
    on=None,
    left_on=['molecule_name','atom_index'],
    right_on=['molecule_name','atom_index'],
    left_index=False,
    right_index=False,
    sort=True,
    suffixes=("_x", "_y"),
    copy=True,
    indicator=False,
    validate=None,
)
magnetic_sheilding_tensors_mulliken_charges_structures_df.to_csv(os.path.join('..','data','joined_data','magnetic_sheilding_tensors_mulliken_charges_structures_df.csv'), index=False)
```

```python
# magnetic_sheilding_tensors_mulliken_charges_structures_df
```

```python
magnetic_sheilding_tensors_mulliken_charges_structures_diapole_moments_potential_energy_df = pd.merge(
    left=magnetic_sheilding_tensors_mulliken_charges_structures_df,
    right=diapole_moments_potential_energy_df,
    how="left",
    on=None,
    left_on=['molecule_name'],
    right_on=['molecule_name'],
    left_index=False,
    right_index=False,
    sort=True,
    suffixes=("_x", "_y"),
    copy=True,
    indicator=False,
    validate=None,
)
magnetic_sheilding_tensors_mulliken_charges_structures_diapole_moments_potential_energy_df.to_csv(os.path.join('..','data','joined_data','magnetic_sheilding_tensors_mulliken_charges_structures_diapole_moments_potential_energy_df.csv'), index=False)
```
